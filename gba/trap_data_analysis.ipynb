```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import os,sys
from time import time
from PIL import Image
from scipy.ndimage import rotate
from scipy.optimize import curve_fit
from numpy import asarray,copy,nan
from IPython.core.display import Image as Im

imdir = './paper_data/'
sys.path.append(os.path.join('.',imdir))
```

```python
files = ['2021-12-29-100748 trap image (iris opened).png',
         '2021-12-29-100748 trap image.png']
ims = [asarray(Image.open(os.path.join(imdir,f)).convert('L'))
       for f in files]
```

```python
opn_im,cls_im = ims
```

```python
fig,ax = plt.subplots(dpi=100)
i2xy = rotate(cls_im,-2.5)[20:630,100:760]
ax.imshow(i2xy)
ax.axhline(536,color='r')
ax.axvline(341)
```

```python
plt.plot(i2xy[536,386:407]/np.amax(i2xy[536,295:407]))
plt.ylim(0,1)
```

```python
# dark trap fit, single site: x**alpha power law fit.
def model(x,x0,a,alpha,b):
    return a*abs(x-x0)**alpha + b

xn = 2*18 # `18 is roughly the pixel separtion between sites, gotten from the Gaussian fits below
linedata = i2xy[536,388-xn:405-xn]#295:407]
linedata = linedata/np.amax(linedata)
xpts = np.linspace(0,len(linedata)-1,len(linedata))

plt.plot(xpts,linedata)
plt.plot(xpts,model(xpts,
                    x0=10,#102, # bounds [6,12]
                    a=0.001, # bounds [0.001,0.01]
                    alpha=2.5, # bounds [1,4]
                    b=0.3)) # bounds [0.1,0.5]
```

```python
um_pixels = 3.45 # um per pixels on ThorCam

sites = 5
fig,axes = plt.subplots(nrows=1,ncols=sites)
fig.tight_layout()

for i in range(sites):

    xn = 18*i
    linedata = i2xy[536,388-xn:405-xn]
    linedata = linedata/np.amax(linedata)
    xpts = np.linspace(0,len(linedata)-1,len(linedata))
    
    popt, pcov = curve_fit(model,xpts,linedata,
                           bounds=([4,0.001,1,0.1],
                                   [14,0.01,5,0.5]))
    x0,a,alpha,b = popt
    axes[i].plot(xpts,linedata,label='data')
    axes[i].plot(xpts,model(xpts,x0,a,alpha,b),ls='--',label='fit')
    axes[i].legend()
    axes[i].set_title(rf"$\alpha$={alpha:.2f}")
```

```python
# dark trap fit: |1 - sum(Gaussian*kroneckerdelta(x-xn), n)|^2
gaussian = lambda a,x,x0,w0: a*np.exp(-((x-x0)/w0)**2)
def model(x,x0,w0,d,a,b,sites):
    field = 1
    for n in range(int(sites+0.5)):
        field -= gaussian(a,x,x0-d*n,w0)
    return b*field**2

linedata = i2xy[536,386:407]#295:407]
linedata = linedata/np.amax(linedata)
xpts = np.linspace(0,len(linedata)-1,len(linedata))

plt.plot(xpts,linedata)
plt.plot(xpts,model(xpts,
                    x0=10,#102, # bounds [90,110]
                    w0=6, # bounds [5,7]
                    d=7*2.6, # bounds [14,21]
                    a=.6, # bounds [0.5,.7]
                    b=1.3, # bounds [1,2]
                    sites=1)) # bounds [5,7]
```

```python
um_pixels = 3.45 # um per pixels

linedata = i2xy[536,295:407]
linedata = linedata/np.amax(linedata)
xpts = np.linspace(0,len(linedata)-1,len(linedata))

popt, pcov = curve_fit(model,xpts,linedata,
                       bounds=([90,5,14,0.5,1,5],
                               [110,7,21,0.7,2,7]))
x0,w0,d,a,b,sites = popt

plt.plot(xpts,linedata,label='data')
plt.plot(xpts,model(xpts,x0,w0,d,a,b,sites),ls='--',label='fit')
plt.legend()
print(f"d={d*um_pixels:.2f}um, w0={w0*um_pixels:.2f}um, a={a}, b={b}, sites={int(sites+0.5)}")
```

### "continuous" phase mask

```python
mm = 10 # plot resolution of 100 um

## physics and system params
lmbda = 5.5e-4
k = 2*pi/lmbda
w = 5 # grid width in mm
pts = 401
xpts = linspace(-w/2, w/2, pts)
ypts = linspace(-w/2, w/2, pts)
xx,yy = meshgrid(xpts,ypts)
rr = sqrt(xx**2 + yy**2)
phi = arctan2(yy,xx)

sample = zeros((pts,pts))

# will be Fresnel zone plate profile after we binarize it

gsteps = [10**x for x in range(0,4)]
ncols = len(gsteps)
fig,axes = plt.subplots(ncols=ncols)
fig.dpi=100
for g,ax in zip(gsteps,axes):
    ax.imshow(0.5*(1+cos(g*rr**2)**2))
    ax.set_title(f'g={g:.1e}')
plt.show()
```

```python
f = 200 # 1st order focal length in mm
g = pi/(2*f*lmbda + 0.5*lmbda**2)
mask = 0.5*(1+cos(g*rr**2)**2)

# plt.imshow(mask)

zsteps=[180,190,200,210,220]
ncols = len(zsteps)
fig,axes = plt.subplots(ncols=ncols)
fig.dpi=150
for z,ax in zip(zsteps,axes):
    #far field
    output,xx2 = propagate(z, mask, k, xx, rr, padding=pts//2, padval=0, logging=True)
    ax.imshow(abs(output)**2/amax(abs(output)**2))#,extent=[xx2[0],xx2[-1],xx2[0],xx2[-1]])
    ax.set_title(f"z={z}mm")
    ax.set_xlim((pts//3,pts-pts//3))
    ax.set_ylim((pts//3,pts-pts//3))
```

## s and p polarization transformation
I use the definitions given in "Fast focus field calculations" by Leutenegger et. al.

x and y are transverse coordinates with x along the vertical direction, and z is in the forward propagation direction. the polarizations defined below are therefore vectors in this Cartesian system. 

```python
ep = lambda phi: array([cos(phi), sin(phi), 0])
es = lambda phi: array([-sin(phi), cos(phi), 0])
er = lambda theta,phi: array([
    cos(phi)*cos(theta),
    sin(phi)*cos(theta),
    sin(theta)])

# axicon
diameter = 2*mm
alpha = 17*pi/180 # the axicon angle from the transverse plane
n_axicon = 2.5295
n = 1

pts = 512
xlist = linspace(-diameter/2,diameter/2,pts)
theta = arcsin(n_axicon*sin(alpha)) - alpha # the same value everywhere
xx,yy = meshgrid(xlist, xlist)
phi = arctan2(yy,xx)
```

```python
# decomposition of linear x-polarized light into ep and es
ex = array([1,0,0])
ax = plt.imshow(dot(ep(phi),ex))
plt.colorbar(ax)
plt.title("ep.ex")
plt.show()
ax = plt.imshow(dot(es(phi),ex))
plt.colorbar(ax)
plt.title("es.ex")
plt.show()
```

```python
# upon transmission through the rotationally symmetric refracting surface,
# the polarization component ep becomes er, which has a small ez component
ez = array([0,0,1])
ax = plt.imshow(dot(er(theta,phi),ex))
plt.colorbar(ax)
plt.title("er.ex")
plt.show()
ax = plt.imshow(dot(er(theta,phi),ex)*(dot(er(theta,phi),ez)))
plt.colorbar(ax)
plt.title("(er.ex)*(er.ez) = how much ex now projects on ez")
plt.show()
ax = plt.imshow(dot(es(phi),ex))
plt.colorbar(ax)
plt.title("es.ex")
plt.show()
```

### high NA 
fails to produce diffraction limited spot size for input grid size up to 4096 wide, but both RS and CZT remain in agreement

```python
diameter = 2 * mm
focal = sqrt(3)*(diameter/2)

input_samples = 2**13
x0 = np.linspace(-diameter / 2, diameter / 2, input_samples)
y0 = np.linspace(-diameter / 2, diameter / 2, input_samples)
wavelength = 0.6238 * um
k = 2*pi/wavelength
```

```python
2**12
```

```python
NA = sin(arctan((diameter/2)/focal))
print("NA = ",NA)
```

```python
u0 = Scalar_source_XY(x=x0, y=y0, wavelength=wavelength)
u0.plane_wave()

t0 = Scalar_mask_XY(x=x0, y=y0, wavelength=wavelength)
t0.lens(r0=(0 * um, 0 * um),
        radius=(diameter / 2, diameter / 2),
        focal=(focal, focal))

t0.draw('phase')

u1 = u0 * t0
```

```python
plt.imshow(abs(u1.u)**2)
```

```python
xmin = -10
xmax = -xmin
ymin = xmin
ymax = xmax
samples=512
```

```python
%%time
N = input_samples
xspan = diameter
dx = (xspan/N)
# define k space grid and field
dfx = 1/(xspan)
fx_list = arange(-N//2,N//2,1)*dfx
kx_list = 2*pi*fx_list
ky_list = -kx_list
KX, KY = meshgrid(kx_list, ky_list)
KZ_real = real(sqrt(k**2 - KX**2 - KY**2 + 0j))
KZ_imag = imag(sqrt(k**2 - KX**2 - KY**2 + 0j))
KZ = KZ_real + 1j*abs(KZ_imag)
u1_kspace = fftshift(fft2(u1.u))

# propagate
z_distance = focal
phase = f*KZ_real % 2*pi
u2_kspace = u1_kspace*exp(1j*z_distance*KZ)

# transform back to real space
u2 = ifft2(u2_kspace)
int_asm = real(conjugate(u2)*u2)
int_asm /= amax(u2)
```

```python
u2_asm = Scalar_source_XY(x=x0, y=y0, wavelength=wavelength)
u2_asm.u = u2
u2_asm.cut_resample(x_limits=(xmin, xmax),
                   y_limits=(ymin, ymax),
                   num_points=(samples,samples),
                   new_field=False,
                   interp_kind=(3, 1))
u2_asm.draw(logarithm=False)
```

```python
dx = (diameter/input_samples)/um
assert wavelength/(2*dx) < 1, wavelength/(2*dx)
```

```python
%%time
u2_rs = u1.RS(z=focal, verbose=True)

u2_rs.cut_resample(x_limits=(xmin, xmax),
                   y_limits=(ymin, ymax),
                   num_points=(samples,samples),
                   new_field=False,
                   interp_kind=(3, 1))
u2_rs.draw(logarithm=False)
```

```python
%%time
xout=np.linspace(xmin,xmax,samples)
yout=np.linspace(ymin,ymax,samples)
u2_czt = u1.CZT(z=focal, xout=xout, yout=yout,
               verbose=True)

u2_czt.draw(logarithm=False) #logarithm=1e-2)
```

```python
%%time
xout=np.linspace(xmin,xmax,samples)
yout=np.linspace(ymin,ymax,samples)
u2_czt = u1.CZT(z=focal, xout=xout, yout=yout,
               verbose=True)

u2_czt.draw(logarithm=False) #logarithm=1e-2)
```

```python
u2_czt.draw_profile(point1=(xmin,0),point2=(xmax,0))
plt.show()
```

```python
# analytical result
a = diameter/2
k = 2*np.pi/wavelength
f = focal
int_func = lambda x: (j1(a*x*k/f)/x)**2
int_analytic = int_func(xout)
int_analytic /= max(int_analytic)

int_czt = abs(u2_czt.u[len(xout)//2])**2
norm = max(int_czt)
int_czt /= norm

int_asm = abs(u2_asm.u[len(xout)//2])**2
norm = max(int_asm)
int_asm /= norm

int_rs = abs(u2_rs.u[len(xout)//2])**2
norm = max(int_rs)
int_rs /= norm

fig,ax = plt.subplots(dpi=300)
ax.plot(xout,int_analytic,label='analytic',c='k')
ax.plot(xout,int_czt,'--',label='CZT',c='orange')
ax.scatter(xout,int_rs,label='RS',c='limegreen',s=2)
ax.scatter(xout,int_asm,label='ASM',c='purple',s=4,marker="P")
ax.set_xlim((-5,5))
plt.axvline(-1.22*wavelength/(2*NA),label='Airy 1st minima')
plt.axvline(1.22*wavelength/(2*NA))
plt.legend()
```
